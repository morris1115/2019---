```python
import pygame
import random

WHITE = (255,255,255)
WINDOW_WIDTH = 1024
WINDOW_HEIGHT = 512
AIRC_WIDTH = 100
AIRC_HEIGHT = 80
    
def drawing(obj,x,y):
    global window
    window.blit(obj,(x,y))

def runGame(): # 게임의 동작을 구현할 함수
    global window, clock, aircraft, background, background_next # 전역변수 선언
    global monster, fires, bullet
    
    bullet_xy = []
    
    x = WINDOW_WIDTH * 0.05 # 비행기 초기 x좌표 위치
    y = WINDOW_HEIGHT * 0.8 # 비행기 초기 높이 x좌표 위치
    y_change = 0 # 변한 높이 값
    
    background_x = 0 #움직이는 화면 0으로 초기설정
    background_x_next = WINDOW_WIDTH 
    
    monster_x = WINDOW_WIDTH
    monster_y = random.randrange(0,WINDOW_HEIGHT) #0부터 높이까지의 랜덤값을 지정
    
    fire_x = WINDOW_WIDTH
    fire_y = random.randrange(0,WINDOW_HEIGHT)
    random.shuffle(fires) #fires리스트를 랜덤하게 섞는다.
    fire = fires[0] #섞인 리스트의 0번째 값을 추출.
    
    
    crashed = False
    while not crashed: #충돌 전 까지 반복
        for event in pygame.event.get(): #어떠한 이벤트가 있을 경우 반복
            if event.type == pygame.QUIT:
                crashed = True # crashed 하면 게임 종료
            if event.type == pygame.KEYDOWN: # 키패드를 눌렀을 때,
                if event.key == pygame.K_UP:
                    y_change = -5
                elif event.key == pygame.K_DOWN:
                        y_change = 5
                elif event.key == pygame.K_SPACE:
                    bullet_x = x + AIRC_WIDTH
                    bullet_y = y + AIRC_HEIGHT / 2
                    bullet_xy.append([bullet_x,bullet_y])
                    
            if event.type == pygame.KEYUP:
                if event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                    y_change=0
        
        y += y_change               
        window.fill(WHITE)
        
        background_x -= 2
        background_x_next -= 2
        
        if background_x <= -WINDOW_WIDTH:
            background_x = WINDOW_WIDTH
        
        if background_x_next <= -WINDOW_WIDTH:
            background_x_next = WINDOW_WIDTH
        
        drawing(background,background_x,0)
        drawing(background_next,background_x_next,0)
        
        if y < 0:
            y = 0
        elif y > WINDOW_HEIGHT - AIRC_HEIGHT:
            y = WINDOW_HEIGHT - AIRC_HEIGHT
        
        drawing(aircraft,x,y)
        
        monster_x -= 7
        if monster_x <= 0:
            monster_x = WINDOW_WIDTH
            monster_y = random.randrange(0,WINDOW_HEIGHT)
        
        if fire is None:
            fire_x -= 30
        else:
            fire_x -= 10
            
        if fire_x <= 0:
            fire_x = WINDOW_WIDTH
            fire_y = random.randrange(0,WINDOW_HEIGHT)
            random.shuffle(fires)
            fire = fires[0]
            
        drawing(monster,monster_x, monster_y)
        
        if fire != None:
            drawing(fire,fire_x,fire_y)
            
        if len(bullet_xy) != 0:
            for i,bxy in enumerate(bullet_xy):
                bxy[0] += 15
                bullet_xy[i][0] = bxy[0]
                if bxy[0] >= WINDOW_WIDTH:
                    bullet_xy.remove(bxy)
        
        if len(bullet_xy) != 0:
            for bx,by in bullet_xy:
                drawing(bullet,bx,by)
        
        pygame.display.update()
        clock.tick(60) #초당 60프레임
        
    pygame.quit()
    quit()

    
def initGame(): 
    global window, clock, aircraft, background, background_next
    global monster, fires, bullet
    
    pygame.init()
    window = pygame.display.set_mode(size = (WINDOW_WIDTH, WINDOW_HEIGHT)) #윈도우 크기 설정
    pygame.display.set_caption("FlyGame") #윈도우 창 상단의 제목 지정
    
    aircraft = pygame.image.load('./img/airplane2.png') #비행기 이미지 불러오기
    aircraft = pygame.transform.scale(aircraft, (AIRC_WIDTH, AIRC_HEIGHT)) # 120x80 픽셀로 비행기 크기 조정
    
    background = pygame.image.load('./img/background.jpg') #배경 이미지 불러오기
    background = pygame.transform.scale(background,(WINDOW_WIDTH, WINDOW_HEIGHT)) # 배경의 크기를 window의 크기만큼 설정
    background_next = background.copy()
    
    monster = pygame.transform.scale(pygame.image.load('./img/devil.png'),(100,70))
    fire1 = pygame.transform.scale(pygame.image.load('./img/fire1.png'),(100,70))
    fire2 = pygame.transform.scale(pygame.image.load('./img/fire2.png'),(100,70))
    
    fires = [fire1, fire2, None, None, None, None, None]
    
    bullet = pygame.transform.scale(pygame.image.load('./img/bullet.png'),(30,20))
    
    clock = pygame.time.Clock() #초당 프레임 초기값 설정
    runGame() #게임을 시작하겠다는 함수호출

if __name__ == '__main__':
    initGame()
```
